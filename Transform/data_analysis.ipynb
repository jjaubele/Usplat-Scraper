## Extraer organizaciones y paises

```python
colegios = atletas["Colegio"].dropna().drop_duplicates()
colegios.index = np.arange(1, len(colegios) + 1)
colegios = pd.DataFrame(colegios)
colegios.to_csv('colegios.csv', index_label='id_colegio')
```

```python
clubes = atletas["Club"].dropna().drop_duplicates()
clubes.index = np.arange(1, len(clubes) + 1)
clubes = pd.DataFrame(clubes)
clubes.to_csv('clubes.csv', index_label='id_club')
```

```python
clubes_master = atletas["Club Master"].dropna().drop_duplicates()
clubes_master.index = np.arange(1, len(clubes_master) + 1)
clubes_master = pd.DataFrame(clubes_master)
clubes_master.to_csv('clubes_master.csv', index_label='id_club_master')
```

```python
paises = atletas["Pais"].dropna().drop_duplicates()
paises.index = np.arange(1, len(paises) + 1)
paises = pd.DataFrame(paises)
paises.to_csv('paises.csv', index_label='id_pais')
```
