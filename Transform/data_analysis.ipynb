```python
import numpy as np
import pandas as pd
```

```python
atletas = pd.read_csv('../BD/Tablas/atletas.csv')
resultados = pd.read_csv('../BD/Tablas/resultados.csv')
pruebas = pd.read_csv('../BD/Tablas/pruebas.csv')
campeonatos = pd.read_csv('../BD/Tablas/campeonatos.csv')
eventos = pd.read_csv('../BD/Tablas/eventos.csv')
```

```python
# Ranking 800 Metros Planos Hombres 2024
ranking_800m_hombres_2024 = resultados.merge(atletas, on='id_atleta').merge(pruebas, on='id_prueba')
ranking_800m_hombres_2024 = ranking_800m_hombres_2024[(ranking_800m_hombres_2024['nombre'] == '800 Metros Planos') & (ranking_800m_hombres_2024['genero'] == 'Hombres') & (ranking_800m_hombres_2024['fecha'] >= '2024-01-01') & (ranking_800m_hombres_2024['fecha'] <= '2024-12-31')]
ranking_800m_hombres_2024 = ranking_800m_hombres_2024.sort_values(by='puntos', ascending=False).reset_index(drop=True)
ranking_800m_hombres_2024.head(20)
```

## Extraer organizaciones y paises

```python
colegios = atletas["Colegio"].dropna().drop_duplicates()
colegios.index = np.arange(1, len(colegios) + 1)
colegios = pd.DataFrame(colegios)
colegios.to_csv('colegios.csv', index_label='id_colegio')
```

```python
clubes = atletas["Club"].dropna().drop_duplicates()
clubes.index = np.arange(1, len(clubes) + 1)
clubes = pd.DataFrame(clubes)
clubes.to_csv('clubes.csv', index_label='id_club')
```

```python
clubes_master = atletas["Club Master"].dropna().drop_duplicates()
clubes_master.index = np.arange(1, len(clubes_master) + 1)
clubes_master = pd.DataFrame(clubes_master)
clubes_master.to_csv('clubes_master.csv', index_label='id_club_master')
```

```python
paises = atletas["Pais"].dropna().drop_duplicates()
paises.index = np.arange(1, len(paises) + 1)
paises = pd.DataFrame(paises)
paises.to_csv('paises.csv', index_label='id_pais')
```
